```python
# Create a dictionary from the dataframe
keys =  df2['Name'].to_list()
values = df2['Letter'].to_list()
dictionary = dict(zip(keys, values))
```

```python
dictionary
```

```python
# Go through the cols E1 - E6 and replace transliteration to Arabic Letter
def create_root_column(col):
    new_col = []
    for row in col:
        if row != 0:
            new_col.append(dictionary[row])
        else:
            new_col.append(0)
    return new_col

df1["R1"] = create_root_column(df1["E1"])
df1["R2"] = create_root_column(df1["E2"])
df1["R3"] = create_root_column(df1["E3"])
df1["R4"] = create_root_column(df1["E4"])
df1["R5"] = create_root_column(df1["E5"])
df1["R6"] = create_root_column(df1["E6"])
```

```python
df1.head(20)
```

```python
missing_value_columns = ['E1', 'E2', 'E3', 'E4', 'E5', 'E6', 'R1', 'R2', 'R3', 'R4', 'R5', 'R6']
for col in df1.columns:
    if col in missing_value_columns:
        df1[col] = df1[col].astype("string")
```

```python
# Create a new column with root letters combined / concatenated into 1 string 
df1['Root Letters'] = df1['R1'] + df1['R2'] + df1['R3'] + df1['R4'] + df1['R5'] + df1['R6'] 
```

```python
df1.head(20)
```

```python
new_col = []
for row in df1['Root Letters']:
    removed_zeros = row.strip("0")
    new_str = ''
    for char in removed_zeros:
        new_str += char
        new_str += ' '
        
    new_col.append(new_str)
```

```python
df1.drop(['Root Letters'], axis=1, inplace=True)
df1["Root_Letters"] = new_col
df1.head()
```

```python
# Export the cleaned data to an excel file
df1.to_excel("/Users/tahir/Desktop/Github/documents/roots/Data/Cleaned_Root_letters.xlsx")
```
